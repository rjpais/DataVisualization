```python
# VENN  DIAGRAM EXAMPLES 
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
from matplotlib_venn import venn3


# Define sets element number and respective intersepts 
A = 250
B = 300 
C = 400
AB = 120
AC = 90
BC = 150
ABC = 55

# Calculated elements with no interseptions (representative elements) 
A1 = A - AB - AC + ABC
B2 = B - AB - BC + ABC
C3 = C - BC - AC + ABC

# Defining set names 
L1 = "A"
L2 = "B"
L3 = "C"

# ploting the Ven diagrams for 2 sets interseption A & B
print("EXAMPLE OF A 2 SET intersept VenDiagram")

venn2(subsets = (A, B , AB), set_labels = (L1, L2))
plt.show()


# ploting the Ven diagrams for 2 3 sets interseption A & B & C 
print("\n\n\n")
print("EXAMPLE OF A 2 SET intersept VenDiagram")

venn3(subsets = (A1, B2, C3, AB, AC, BC , ABC), set_labels = (L1 , L2, L3) )
plt.show() 
```
