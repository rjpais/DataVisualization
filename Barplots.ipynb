```python
# EXAMPLE OF HOW TO MAKE A SIMPLE BAR PLOT

import matplotlib.pyplot as plt
    
# use a predifined style of plots ( this case ggplots from R)

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (20, 10)


X = [ "X1", "X2",  "X3", "X4", "X5" ,  "X6" ] # variable x
Y  = [ 1, 4,  2 , 6, 3 ,  5 ]      # variable y  


# title of plot
plt.title("add title of plot", size = 20 )

# axis lables 
plt.xlabel("\n X variable name", size = 20 )
plt.ylabel("Y variable name \n", size = 20 )
plt.tick_params(axis='both', labelsize = 20) # set the font of axis values

# plot and show data
plt.bar(X, Y, color = "purple" )
plt.show()  
```

```python
# EXAMPLE OF HOW TO MAKE A BAR PLOT WITH LEGEND OF MULTIPLE CLASSES WITH DIFFERENT COLORS

import matplotlib.pyplot as plt
    
# use a predifined style of plots ( this case ggplots from R)

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (20, 10)


Class1 = [ "X1", "X2" ] # variable x
Class2 = [ "X3", "X4", "X5" ] # variable x
Class3 = [ "X6" ] # variable x
YClass1  = [ 1,  4 ]      # variable y of class 1  
YClass2 = [ 2 , 6, 3 ]      # variable y of class 2 
YClass3 = [ 5 ]      # variable y of class 3 


# title of plot
plt.title("add title of plot", size = 20 )

# axis lables 
plt.xlabel("\n X variable name", size = 20 )
plt.ylabel("Y variable name \n", size = 20 )
plt.tick_params(axis='both', labelsize = 20) # set the font of axis values

# plot and show data
plt.bar(Class1, YClass1, color = "orange", label = "Class 1")
plt.bar(Class2, YClass2, color = "blue", label =  "Class 2" )
plt.bar(Class3, YClass3, color = "green", label =  "Class 3") 
plt.legend(fontsize = 18 )
plt.show()  
```

```python
# EXAMPLE OF A GROUPED BARPLOT

import matplotlib
import matplotlib.pyplot as plt
import numpy as np


    
# use a predifined style of plots ( this case ggplots from R)
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (20, 10)

labels = ['X1', 'X2', 'X3', 'X4', 'X5','X6' ]
Y1 = [5, 6, 9, 3, 4, 1]
Y2 = [3, 7, 3, 5, 2, 2]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, Y1, width, label='Group 1')
rects2 = ax.bar(x + width/2, Y2, width, label='Group 2')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Y variable' , size = 20 )
ax.set_title('Add a title here', size =  20 )
ax.set_xticks(x )
ax.set_xticklabels(labels, size = 20)
ax.legend( fontsize = 18)
plt.tick_params(axis='both', labelsize = 20) # set the font of axis values


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

```python
# BARPLOT TRANSPARENT AND ROTATED  


import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import math 

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (20, 10)


NAMES = [ "SAMPLE " + str(n+1) for n in range(8)  ]
POSITIONS = [i for i in range(len(NAMES))]
VALUES = [ 18264, 15486 , 4327 ,12899 ,10670 , 10010, 855 ,4327]

plt.bar(POSITIONS, VALUES, color = "orange", alpha=0.5)
for i, value in enumerate(VALUES):
    iText = str(value)
    TextPosition = i   
    plt.text( TextPosition, 500 , iText , horizontalalignment='center',
        fontsize=16 , rotation=90 ,verticalalignment='center')

plt.ylabel('Y name \n' , size = 20)
plt.tick_params(axis='both', labelsize = 20) # set the font of axis values
plt.yscale('symlog')
plt.xticks(POSITIONS , NAMES , rotation = 90)
plt.yticks(rotation = 90)
plt.ylim([0,1000000 ])
plt.show() 
    
```
