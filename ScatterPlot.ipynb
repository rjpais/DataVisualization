```python
# MULTIPLE SCATER PLOT TYPES AND OVERLAPING WITH INTERCEPT  


# plot of curves using python with ggplots style 
import matplotlib.pyplot as plt


# use a predifined style of plots ( this case ggplots from R)
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (20, 10)


# lists of values of variables 
X = [ xi for xi in range(0, 100) ] # independent variable x 
Y = [ float (200*xi^2) / ( xi  + 20)  for xi in X ] # Hyperbolic function (dependent variable)
K  = [ 100  for xi in X ]      # constant function  (dependent variable)
Z = [ 3*xi + 3  for xi in X ]  # linear function  (dependent variable)

# a way of intercepting functions 
I = [xi for i, xi in enumerate(X) if round(Y[i], 0) == round(Z[i], 0) ] # get the maching values of Y and Z functions   
print ("intercept Y with Z : x  = ", max(I) )


# title of plot
plt.title("intercept of functions")

# axis lables 
plt.xlabel("x-axis", size  =  20)
plt.ylabel("y-axis" , size  =  20 )
plt.tick_params(axis='both', labelsize = 20) # set the font of axis values

# some ploting examples and show data
plt.plot(X, Z, color = "red", label = "linear", linestyle = "--",  linewidth=6 )
plt.plot(X, Y, color = "blue", label = "hyperbolic" , linewidth = 6 )
plt.plot(X, K, "k--", label = "constant", linestyle = ":" , linewidth=6 )
plt.scatter(X, Y, color = "black", marker = ".", s = 100  )
plt.legend(fontsize = 18 )
plt.show()  
```

```python
# PLOT CURVES WITH FILLING COLOR IN BETWEEN 
import matplotlib.pyplot as plt

# use a predifined style of plots ( this case ggplots from R)
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (20, 10)

Curve1 = [ x**2 - 10*x - 1000 for x in range(100) ]
Curve2 = [ 2*x  + 10  for x in range(100) ]
Curve3 = [ -4*x**2 + x - 1000 for x in range(100) ]

plt.xlabel('Y', size = 20 )
plt.ylabel('X', size =20 )
plt.tick_params(axis='both', labelsize = 18) # set the font of axis values
plt.plot( [x for x in range(100) ], Curve1, "r--", label = "curve 1" , linewidth = 4 )
plt.plot([x for x in range(100) ], Curve2, "b--", label = "curve 2" , linewidth = 4 )
plt.plot([x for x in range(100) ], Curve3, "g--", label = "curve 3" , linewidth = 4 )

plt.fill_between([x for x in range(100) ], Curve1,  Curve2 ,
                 facecolor="yellow", # The fill color
                 color='orange',       # The outline color
                 alpha=0.3, label = "up area" )     
plt.fill_between([x for x in range(100) ], Curve2,  Curve3 ,
                 facecolor="purple", # The fill color
                 color='blue',       # The outline color
                 alpha=0.3, label = "bottom area" )  
plt.legend( fontsize = 18, loc = "lower left")
plt.show()

```

```python

# ERROR BARS EXAMPLE 
import matplotlib.pyplot as plt
import numpy as np 

# Plot the table values with bar stander deviations   
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (20, 10)
plt.tick_params(axis='both', labelsize = 18) # set the font of axis values
plt.xlabel('Y', size = 20 )
plt.ylabel('X', size =20 )



# X-axis elements names 
Xelements = ["ID" + str(n+1) for n in range(10)] 
# list with average values 
AVG1 = [3, 4, 2, 5, 6, 2, 8 , 5 , 4 , 7] 
AVG2 = [ x*0.7 for x in AVG1 ] 
# standard deviations of elements
STDEV1 = [0.3 , 0.4, 0.5, 0.3, 0.5, 0.2, 0.4, 0.5, 0.3, 0.6 ] 
STDEV2 = [ x -0.05 for x in STDEV1 ] 


plt.errorbar( [i + 0.1 for i in range(10) ],
             AVG1 ,  
            STDEV1 , 
             linestyle='None', marker='s', label = "E1", color ="blue"  )
         
plt.errorbar( [i - 0.1 for i in range(10) ] ,
             AVG2,  
            STDEV2 , linestyle='None', marker='o', 
             label = "E2", color = "red" ) 
plt.xticks(np.arange(10), Xelements,
           fontsize=18, rotation = 90 )
plt.legend( fontsize = 18, loc ="upper left" )

plt.show()



```
